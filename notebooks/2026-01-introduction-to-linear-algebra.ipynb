### 2. Motivation for Linear Algebra

- Linear algebra is the **language of machine learning**
- It allows representing data as vectors and matrices
- Enables transformations of datasets and model parameters
- Essential foundation for ML and Data Engineering

### 3. Vectors Concepts

- Vectors have **magnitude** and **direction**
- Represent points in space or features in datasets
- Key operations:
    - Addition
    - Subtraction
    - Scaling (multiplying by a scalar)

- Exploring parameter space:
    - Parameters define model behavior
    - Visualizing vectors helps understand geometric relationships

### 4. Solving simultaneous Equations

- Systems of equations illustrate relationships between variables
- Foundation for regression, ML models and data pipelines
- Understanding solution strategies conceptually

### 5. Vector Operations

- Conceptual exercises:
    - v1 * v2 moves the tip of a vector
    - 3 * v1 scales magnitude

- Understanding operations prepares for feature transformations
